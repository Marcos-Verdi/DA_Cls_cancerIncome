## Which states have counties on a falling trend?

```python
falling_trend = df.loc[df.recentTrend == 'falling']
```

```python
top_falling_states = falling_trend.groupby('State').count().sort_values('zipCode',ascending=False).head(15)
top_falling_states
```

```python
plt.figure(figsize=(20,10))
plt.bar(x=top_falling_states.index,height=top_falling_states['zipCode'])
plt.xlabel('States')
plt.title('States with most counties on a falling trend')
```

We see that the states with more counties with a falling trend are California, Texas, New York, Florida, Illinois, New Jersey, and Massachusetts. 

```python
rising_trend = df.loc[df.recentTrend == 'rising']
```

```python
top_rising_states = rising_trend.groupby('State').count().sort_values('zipCode',ascending=False).head(15)
top_rising_states
```

```python
plt.figure(figsize=(20,10))
plt.bar(x=top_rising_states.index,height=top_rising_states['zipCode'])
plt.xlabel('States')
plt.title('States with most counties on a rising trend')
```

The states with more counties on a rising trend are Iowa, New Jersey, Lousiana, Utah, and New Mexico.

## Five Year Trend Analysis

```python
df['fiveYearTrend']=df['fiveYearTrend'].apply(lambda x: np.nan if x=='*' else x)
```

```python
fiveY_trend_df = df.copy(deep=True)
fiveY_trend_df
```

```python
fiveY_trend_df = fiveY_trend_df.dropna()
fiveY_trend_df.head()
```

```python
fiveY_trend_df.info()
```

```python
fiveY_trend_df['fiveYearTrend'] = fiveY_trend_df['fiveYearTrend'].apply(lambda x:pd.to_numeric(x))
```

```python
fiveY_trend_df.info()
```

```python
fiveY_trend_df['County_State'] = fiveY_trend_df['Name'] + '_' + fiveY_trend_df['State']
```

```python
lowest_fiveYtrend = fiveY_trend_df.groupby('County_State').mean().sort_values('fiveYearTrend',ascending=True).head(25)
lowest_fiveYtrend
```

```python
highest_fiveYtrend = fiveY_trend_df.groupby('County_State').mean().sort_values('fiveYearTrend',ascending=False).head(25)
highest_fiveYtrend
```

```python
## Filter by zipcode
```

```python
fiveY_lowest_zip = fiveY_trend_df.sort_values('fiveYearTrend',ascending=True).head(500)
fiveY_lowest_zip
```

```python
plt.figure(figsize=(20,10))
sns.countplot(fiveY_lowest_zip['State'])
```

The state with most zipcodes with a low (negative) rate for the last five years are Virginia, Oklahoma, Texas, Ohio, and Montana. It must be noted that we picked to top 500 zipcodes with the highest *negative rates* 

```python
fiveY_highest_zip = fiveY_trend_df.sort_values('fiveYearTrend',ascending=False).head(500)
fiveY_highest_zip
```

```python
plt.figure(figsize=(20,10))
sns.countplot(fiveY_highest_zip['State'])
```

The state with most zipcodes with a high trend rate for the last five years are Texas, Virginia, Michigan, Kentucky, and Nevada. It must be noted that we picked from the top 500 zipcodes with the highest *positive rates* 

## Analisis of zipcode rates against their poverty rates

```python
fiveY_highest20_zip = fiveY_trend_df.sort_values('fiveYearTrend',ascending=False).head(50)
high_counties = fiveY_highest20_zip.groupby('County_State').mean()
high_counties = high_counties.sort_values('povertyPercent',ascending=False)
high_counties.povertyPercent
```

```python
fiveY_lowest20_zip = fiveY_trend_df.sort_values('fiveYearTrend',ascending=True).head(50)
low_counties = fiveY_lowest20_zip.groupby('County_State').mean()
low_counties = low_counties.sort_values('povertyPercent',ascending=False)
low_counties.povertyPercent
```

### Lets try to remove some of the outliers which could be influencing our outcome. THey are in the Incidence Rate and Average Deaths per Year

```python
df2 = df.loc[(df.incidenceRate < 1000) & (df.avgDeathsPerYear < 10000)]
df2
```

```python
plt.figure(figsize=(20,10))
sns.regplot(data=df2,x='medIncome',y='incidenceRate')
plt.title('Median Income vs Incidence Rate')
```

```python
plt.figure(figsize=(20,10))
sns.regplot(data=df2,x='povertyPercent',y='incidenceRate')
plt.title('Poverty Percent vs Incidence Rate')
```

```python
plt.figure(figsize=(20,10))
sns.regplot(data=df2,x='medIncome',y='avgDeathsPerYear')
plt.title('Median Income vs Average Deaths per Year')
```

```python
plt.figure(figsize=(20,10))
sns.regplot(data=df2,x='povertyPercent',y='avgDeathsPerYear')
plt.title('Poverty Percent vs Average Deaths per Year')
```

We can observe that by removing the outliers from the Incidence Rate column and the Average Deaths Per Year column we actually see the previous trends reverse to be more in accordance with our other results which indicated that higher income communities experience lower levels of incidences and lower rates.
